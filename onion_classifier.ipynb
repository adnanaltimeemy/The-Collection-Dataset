```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# -----------------------------
# 1. Load / define your data
# -----------------------------
# Example structure (replace with your real dataset load)
# df = pd.read_csv("your_file.csv")

texts = df["text"].astype(str).values
y = df["label"].values

# -----------------------------
# 2. Fit TF-IDF vectorizer
# -----------------------------
tfidf = TfidfVectorizer(
    stop_words="english",   # optional but recommended
    max_df=0.9,
    min_df=2
)

X_tfidf = tfidf.fit_transform(texts)

# -----------------------------
# 3. Get feature names
# -----------------------------
feature_names = tfidf.get_feature_names_out()

# -----------------------------
# 4. Function to get top words per class
# -----------------------------
def get_top_words(class_index, n=10):
    class_docs = X_tfidf[y == class_index]
    
    # mean TF-IDF score per feature
    mean_tfidf = np.asarray(class_docs.mean(axis=0)).flatten()
    
    # top indices
    top_indices = mean_tfidf.argsort()[-n:][::-1]
    
    return [feature_names[i] for i in top_indices]

# -----------------------------
# 5. Output results
# -----------------------------
print("Top words - Not Onion (0):", get_top_words(0))
print("Top words - Onion (1):", get_top_words(1))
```
